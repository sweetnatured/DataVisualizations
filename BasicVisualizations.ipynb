```python
import matplotlib.pyplot as plt
```

```python
plt.plot(["a","b","c"],[10,20,30],label ="Increasing ")
plt.plot(["a","b","c"],[30,20,10],label ="Decreasing")
plt.title("Data Visualization\n")
plt.xlabel("Letters")
plt.ylabel("Numbers")
plt.legend()
```

```python
x= ["a","b","c","d","e","f","g"]
y= [2,4,6,8,7,5,4]
x2=["h","i","j","k","l","m","n"]
y2=[1,3,5,7,4,3,1]
plt.title("Bar\n")
plt.bar(x,y,color='green')
plt.bar(x2,y2,color="black")
plt.legend()
```

```python
import numpy as np
np.random.seed(5)
x=np.random.normal(0,1,10000)
plt.hist(x,bins=10)
```

```python
x=np.random.rand(10)
y=np.random.rand(10)
plt.scatter(x,y,color='green',marker='+',s=90)
```

```python
x=np.random.rand(10)
y=np.random.rand(10)
z=np.random.rand(10)
t=np.random.rand(10)
plt.stackplot(x,y,z,t,colors=['c','r','g','b'])
# INTERESTING RESULTS :)
```
